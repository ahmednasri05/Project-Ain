```python
import time
from google import genai
from google.genai import types

client = genai.Client(api_key=api_key)

# 1. Upload using raw string (r"") to handle Windows backslashes
video_file = client.files.upload(file=r"C:/Users/shels/Documents/wezareit el dakhleya/dakhleyaVideos/women brutality stairs/women brutality stairs.mp4")

# 2. WAIT for the video to be processed (Required)
print(f"Uploading and processing {video_file.name}...")
while video_file.state.name == "PROCESSING":
    time.sleep(2)
    video_file = client.files.get(name=video_file.name)

if video_file.state.name == "FAILED":
    raise ValueError("Video processing failed.")

# 3. Analyze using Gemini 3 structure
response = client.models.generate_content(
    model='gemini-3-flash-preview',
    contents=[
        video_file,
        "analyze the video and give me detailed description of what happens in it"
    ],
    config=types.GenerateContentConfig(
        temperature=1.0, # Recommended for Gemini 3 reasoning models
        # Use types.MediaResolution.HIGH for "Original" quality 
        # Or types.MediaResolution.UNSPECIFIED for "Automatic"
        media_resolution=types.MediaResolution.MEDIA_RESOLUTION_MEDIUM, 
        response_mime_type="application/json"
    )
)

print(response.text)
```
